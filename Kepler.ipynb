```python
from lightkurve import search_lightcurve
import matplotlib.pyplot as plt

# Beispiel-Stern: Kepler-10 (bekannter Planet)
lc = search_lightcurve("Kepler-10", mission="Kepler").download()

# Rohdaten plotten
lc.plot()
plt.title("Kepler-10 Raw Light Curve")
plt.show()
```

```python
# Flatten entfernt Trends/Rauschen
lc_flat = lc.flatten(window_length=401)

# GeglÃ¤ttete Kurve plotten
lc_flat.plot()
plt.title("Flattened Light Curve â€“ Transit becomes visible")
plt.show()
```

```python
# Threshold: alles unter Mittelwert - 0.001 markieren
threshold = lc_flat.flux.mean() - 0.001

plt.figure(figsize=(10,4))
plt.plot(lc_flat.time, lc_flat.flux, label='Flattened Light Curve')
plt.scatter(lc_flat.time[lc_flat.flux < threshold],
            lc_flat.flux[lc_flat.flux < threshold],
            color='red', label='Potential Transits')
plt.xlabel("Time (days)")
plt.ylabel("Normalized Brightness")
plt.title("Transit Detection Example")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# lc_flat ist ein LightCurve-Objekt â€“ wir greifen auf .time.value und .flux.values zu
time = lc_flat.time.value   # in Tagen
flux = lc_flat.flux.value   # normalisierte Helligkeit

# Threshold definieren
threshold = flux.mean() - 0.001

# Plot
plt.figure(figsize=(10,4))
plt.plot(time, flux, label='Flattened Light Curve')
plt.scatter(time[flux < threshold], flux[flux < threshold],
            color='red', label='Potential Transits', s=10)
plt.xlabel("Time (days)")
plt.ylabel("Normalized Brightness")
plt.title("Transit Detection Example")
plt.legend()
plt.show()
```
